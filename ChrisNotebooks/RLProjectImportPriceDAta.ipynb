```python
import os
import pandas as pd

def concatenate_csvs(folder_path):
    all_data = pd.DataFrame()
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".csv"):
                csv_file_path = os.path.join(root, file)
                # Define column data types
                column_types = {'NODE': 'str', 'TYPE': 'str', 'VALUE': 'str',}
                for column in range(1, 25):  # Assuming column 1 to 24 are float values
                    column_types[f'HE{column}'] = 'float64'
                
                # Read CSV file with specified data types and handle thousands comma
                df = pd.read_csv(csv_file_path, dtype=column_types, parse_dates=['MARKET_DAY'], thousands=',')
                all_data = pd.concat([all_data, df], ignore_index=True, sort=False)
    return all_data

# Provide the path to the directory containing the CSV files
folder_path = "misolmps/rt"

# Call the function to concatenate all CSV files
concatenated_data = concatenate_csvs(folder_path)

# Display the concatenated data
print(concatenated_data)
```
