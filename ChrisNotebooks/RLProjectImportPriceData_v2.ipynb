```python

#chat gpt
import os
import zipfile

def unzip_folders_in_place(folder_path):
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".zip"):
                zip_file_path = os.path.join(root, file)
                with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
                    zip_ref.extractall(root)
                os.remove(zip_file_path)

# Provide the path to the directory containing the zipped folders
folder_path = "misolmps/da"

# Call the function to unzip all folders in place
unzip_folders_in_place(folder_path)
```

```python
from typing import List, Optional
# ChatGPT
def process_bad_line(bad_line: List[str]) -> Optional[List[str]]:
    """
    Process a single bad line:
    - Replace all instances of '-1,' with '-1'
    - Replace all instances of ',1,' with ',1'
    - Split the line by commas
    """
    # Join the list to form a string
    line = ','.join(bad_line)
    # Replace '-1,' with '-1' and ',1,' with ',1'
    line = line.replace('-1,', '-1').replace(',1,', ',1')
    # Split the modified line by commas
    processed_line = line.split(',')
    # Return the processed line
    return processed_line
```

```python
import os
import pandas as pd

def concatenate_csvs(folder_path):
    all_data = pd.DataFrame()
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".csv"):
                csv_file_path = os.path.join(root, file)
                # Define column data types
                column_types = {'NODE': str, 'TYPE': str, 'VALUE': str,}
                for column in range(1, 25):  # Assuming column 1 to 24 are float values
                    column_types[f'HE{column}'] = 'float64'
                print(file)
                # Read CSV file with specified data types and handle thousands comma
                df = pd.read_csv(csv_file_path, dtype=column_types, parse_dates=['MARKET_DAY'], thousands=',', on_bad_lines = 'warn')
                # df = pd.read_csv(csv_file_path, dtype=column_types, parse_dates=['MARKET_DAY'], thousands=',', engine='python', on_bad_lines = process_bad_line)
                all_data = pd.concat([all_data, df], ignore_index=True, sort=False)
    return all_data

# Provide the path to the directory containing the CSV files
folder_path = "misolmps/da"

# Call the function to concatenate all CSV files
concatenated_data = concatenate_csvs(folder_path)

# Display the concatenated data
print(concatenated_data)
```
